# old_log_probs vs log_probs

```python
import polars as pl
import plotly.express as px

old_log_probs = pack["loss_data"]["old_log_probs"]
log_probs = pack["loss_data"]["log_probs"]
print(old_log_probs.shape, log_probs.shape)
# print(old_log_probs, log_probs)

df = pl.concat([
    pl.DataFrame({"index": range(len(old_log_probs)), "value": old_log_probs.tolist(), "type": "old_log_probs"}),
    pl.DataFrame({"index": range(len(log_probs)), "value": log_probs.tolist(), "type": "log_probs"}),
])
fig = px.scatter(df, x="index", y="value", color="type", title="old_log_probs vs log_probs")
fig.show()
```

```python
diff = log_probs - old_log_probs
df_diff = pl.DataFrame({"index": range(len(diff)), "value": diff.tolist()})
fig = px.scatter(df_diff, x="index", y="value", title="log_probs - old_log_probs")
fig.show()
```

```python
fig = px.histogram(df_diff, x="value", title="log_probs - old_log_probs (histogram)")
fig.show()
```
